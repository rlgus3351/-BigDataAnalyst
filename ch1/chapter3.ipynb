### 6. 필터링,분산

#### 문제
- 1. 'f3' 컬럼이 gold 이면서 'f2'컬럼이 2인 데이터를 찾으시오.
- 2. 찾은 데이터 에서 'f1'컬럼의 분산을 구하시오.

### 7. 필터링,분산

#### 문제
- 1. 모든 나이(age)에 1을 더하시오.
- 2. 20대의 'views' 평균과 30대의 'views' 차이의 절대값을 구하시오


```python
import pandas as pd
df = pd.read_csv('./data/type1_data1.csv')

df['age']=df['age']+1

cond1 = (df['age']>=20) & (df['age']<30)
cond2 = (df['age']>=30) & (df['age']<40)

result = abs(df[cond1]['views'].mean()) - abs(df[cond2]['views'].mean())

print(round(result,2))
```

### 16. 이상치 ,IQR
#### 문제
- 1. views 컬럼의 1사분위수, 3분위수 그리고 IQ을 계산하시오.
- 2. 이상치 조건에 맞는 데이터를 찾으시오.
- 3. 이상치 데이터의 'views' 컬럼 합을 정수로 구하시오


```python
import pandas as pd
df = pd.read_csv('./data/type1_data1.csv')

q1 = df['views'].quantile(.25)
q3 = df['views'].quantile(.75)
IQR = q3-q1

print(q1,q3,IQR)

line1 = q1-IQR*1.5
line2 = q3+IQR*1.5

cond1 = df['views']<line1
cond2 = df['views']>line2

df = df[cond1 | cond2]
print(int(df['views'].sum()))
```

### 17. 이상치, 소수점 있는 데이터 찾기, 표준 편차
#### 문제
- 1. 'views' 컬럼의 표준편차를 구하시오.
- 2. 'age' 컬럼의 이상치(소수점 나이, 음수 나이, 0살)를 제거하고, 'views' 컬럼의 표준편차를 구하시오,
- 3. 이상치 제거 전후의 'views' 컬럼의 표준편차를 더하여, 반올림 후 소수 둘째 자리까지 구하시오.

```python
import pandas as pd
df = pd.read_csv('./data/type1_data1.csv')

r1 = df['views'].std()

cond = df['age']<=0
df=df[~cond]

cond = df['age'] == round(df['age'],0)
df = df[cond]

r2 =df['views'].std()

print(round(r1+r2,2))
```

### 18. 데이터(행) 기준 평균값 ,인덱싱
#### 문제
- 1. index '2001' 평균보다 큰 값의 개수를 구하시오.
- 2. index '2003' 데이터에서 평균보다 작은 값의 개수를 구하시오.
- 3. 두 개수를 더하시오.

```python
import pandas as pd
df = pd.read_csv('./data/type1_data2.csv',index_col="year")

# 2001, 평균보다 큰 값 합계
m1 = df.loc[2001].mean()
cond1 = df.loc[2001] > m1
r1 = sum(cond1)

m2 = df.loc[2003].mean()
cond2 = df.loc[2003] < m2
r2 = sum(cond2)

print(r1+r2)
```
