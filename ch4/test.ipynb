```python
import pandas as pd

X_tr = pd.read_csv('./ch2_data/X_train.csv')
y_tr = pd.read_csv('./ch2_data/y_train.csv')
X_val = pd.read_csv('./ch2_data/X_test.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
cols = X_tr.select_dtypes('O').columns

for col in cols:
    le = LabelEncoder()
    X_tr[col] = le.fit_transform(X_tr[col])
    X_val[col] = le.transform(X_val[col])
```

```python
X_tr = X_tr.drop('ID',axis=1)
X_test_id = X_val.pop('ID')
```

```python
from sklearn.model_selection import train_test_split
X_tr,X_val,y_tr,y_val = train_test_split(
    X_tr, y_tr['Reached.on.Time_Y.N'],
    test_size = 0.2, random_state=0
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
```

```python
lr = LogisticRegression()
model = lr.fit(X_tr,y_tr)
pred = lr.predict_proba(X_val)

print(roc_auc_score(y_val,pred[:,1]))
```
