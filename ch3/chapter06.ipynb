## Section01 : 단일 표본 검정

### 문제 1.
- 한 커피 제조회사에서는 새로 출시한 커피의 카페인 함량이 평균 95mg 미만으로 주장
- 25개의 커피 샘플을 무작위로 추출하여 커피 제조회사의 주장이 타당한지 검정
- 귀무가설 : u>=95mg
- 대립가설 : u< 95mg


1. 표본 데이터의 평균을 구하시오
2. Shaprio-wilk 검정의 p-value를 구하시오
3. 단일 표본 t-검정의 통계량을 구하시오
4. 단일 표본 t-검정의 p-value를 구하시오
5. 유의수준 0.05이하 귀무가설을 기준으로 검정 결과를 채택/기각 중 선택해 입력


```python
import pandas as pd
df = pd.DataFrame({
    'Caffeine(mg)':[
        94.2,93.7,95.5,93.9,94.0,95.2,94.7,93.5,92.8,94.4,
        93.8,94.6,93.3,95.1,94.3,94.9,93.9,94.8,95.0,94.2,93.7,94.4,95.1,94.0,93.6
    ]
})
```

```python
# 1. 표본 평균
print(df.mean())

# 2. 정규성 검정
from scipy import stats
print(stats.shapiro(df['Caffeine(mg)']))

# 3. 단일 표본 t-검정
print(stats.ttest_1samp(df['Caffeine(mg)'],95,alternative='less'))
```

```python
statistic,pvalue = stats.ttest_1samp(df['Caffeine(mg)'],95,alternative='less')
print("{:.10f}".format(pvalue))
```

```python
# 1. 표본 데이터의 평균을 구하시오
    # -> 94.264
# 2. Shaprio-wilk 검정의 p-value를 구하시오
    # -> 0.9322031137747027
# 3. 단일 표본 t-검정의 통계량을 구하시오
    # ->-5.501737036221897
# 4. 단일 표본 t-검정의 p-value를 구하시오
    # 0.0000058687
# 5. 유의수준 0.05이하 귀무가설을 기준으로 검정 결과를 채택/기각 중 선택해 입력
    # 기각
```

### 문제 2. 
- 스마트폰 배터리의 충전 시간을 줄이기 위한 새로운 충전기 개발
- 스마트폰 사용자 집단을 두 그룹으로 나눠 한 그룹에는 새로운 충전기를 제공하고, 다른 그룹에는 기존의 충전기를 제공
- 두 그룹 충전 시간은 정규 분포를 따르며 분산을 같다고 가정
- 새로운 충전기가 더 빠르다고 할 수 있는지를 유의수준 5%이하에서 검정
---
- 귀무가설 : u1>=u2
- 대립가설 : u1<u2

1. 위의 가설을 검정하기 위한 검정 통계량을 구하시오.
2. 위의 통계량에 대한 p-value를 구하시오.
3. 유의수준 0.05하에서 귀무가설을 기준으로 검정의 결과를 채택/기각 중 선택


```python
import pandas as pd

df = pd.DataFrame({
    '충전기':['New']*10 + ['Old']*10,
    '충전시간':[
        1.5,1.6,1.4,1.7,1.5,1.6,1.7,1.4,1.6,1.5,
        1.7,1.8,1.7,1.9,1.8,1.7,1.8,1.9,1.7,1.6
    ]
})
df
```

```python
from scipy import stats

new_cond = df['충전기']=='New'
old_cond = df['충전기']=='Old'

print(stats.ttest_ind(df[new_cond]['충전시간'],df[old_cond]['충전시간'],alternative='less',equal_var=True))
```

## 문제 4.
- 4개의 중학교에서 다른 교육 방법을 사용해 학생들에게 수학을 가르쳤다.
- 각 중학교의 학생들이 받은 교육 후의 수학 성적은 다음과 같아.
- 4개의 교육방법에 따른 성적 차이가 통계적으로 유의미한지 5%의 유의수준에서 검정.
---
- 귀무가설 : 모든 중학교의 학생들의 수학 성정 평균은 동일하다.
- 대립가설 : 적어도 2개 중학교의 학생들의 수학 성적 평균은 다르다.

1. 각 그룹의 수학 성적에 대해 shaprio-wilk 검정을 통한 정규성을 확인, p-value
2. 4개 그룹의 수학 성정이 등분산성을 갖는기 확인하기위해 levence 검정을 실시 p-value
3. 유의수준 0.05하에서 귀무가설을 기준으로 검정의 결과를 채택/기각 중 선택해 입력
4. 그룹 변수의 자유도
5. 잔차의 자유도
6. 성적의 제곱합
7. 성적의 평균 제곱
8. F-통계량
9. 성적에 대한 p-value

```python
import pandas as pd

df = pd.read_csv('./ch6_data/math.csv')
df['groups'].value_counts()
```

```python
from scipy import stats

# 1. shaprio-wilk 검정
condA = df['groups'] == 'group_A'
print("A",stats.shapiro(df[condA]['scores']))

condB = df['groups'] == 'group_B'
print("B",stats.shapiro(df[condB]['scores']))

condC = df['groups'] == 'group_C'
print("C",stats.shapiro(df[condC]['scores']))

condD = df['groups'] == 'group_D'
print("D",stats.shapiro(df[condD]['scores']))
```

```python
# 2. Levence 검정(등분산성)

print(stats.levene(df[condA]['scores'],df[condB]['scores'],df[condC]['scores'],df[condD]['scores']))
```

```python
# 3. 일원분산분석을 위한 모델 학습
from statsmodels.formula.api import ols

model =ols('scores ~ groups',df).fit()
```

```python
# 3~9 ANOVA 테이블
from statsmodels.stats.anova import anova_lm
print(anova_lm(model))
```

```python
# 1. 각 그룹의 수학 성적에 대해 shaprio-wilk 검정을 통한 정규성을 확인, p-value
    # A  : pvalue=np.float64(0.9051800443853549))
    # B  : pvalue=np.float64(0.6678172590861583))
    # C  : pvalue=np.float64(0.44732595113862045))
    # D  : pvalue=np.float64(0.25824165549017347))
# 2. 4개 그룹의 수학 성정이 등분산성을 갖는기 확인하기위해 levence 검정을 실시 p-value
    # LeveneResult pvalue=np.float64(0.17270284963232108))
# 3. 유의수준 0.05하에서 귀무가설을 기준으로 검정의 결과를 채택/기각 중 선택해 입력
    # 기각
# 4. 그룹 변수의 자유도
    # 3
# 5. 잔차의 자유도
    # 36
# 6. 성적의 제곱합
    # 411.8
# 7. 성적의 평균 제곱
    # 137.26667
# 8. F-통계량
    # 34.174274
    
# 9. 성적에 대한 p-value
    # 1.240642e-10
```

## section 08. 다중 선형 회귀


### 문제 8.
- 한 도시의 배달 서비스 회사에서 한 달 동안의 주문 데이터를 수집했다.
- 배달 주문량은 할인율,온도,광고비와 같은 다양한 요인들에 영향을 받을 수 있다.
- 배달 서비스 회사는 이런 요인들을 고려해 주문량을 예측하고자 한다.
- 할인율, 온도,광고비를 독립변수로 사용해 다중 선형 회귀 모델을 구축하고 다음 문제에 대해 합을 하시오.
---
1. 할인율과 온도의 상관계수를 구하시오
2. 모델의 결정 계수를 구하시오
3. 각 변수의 회귀 계수를 구하시오
4. 모델의 절편을 구하시오
5. 온도의 회귀계수가 통계적으로 유의한지 검정하시오.
6. 데이터가 할인율이 10%, 온도가 20, 광고비가 500,000원일때 배달 주문량을 예측하시오.
7. 잔차 제곱합을 구하시오
8. MSE를 구하시오.
9. 온도의 회귀 계수에 대한 90% 신뢰 구간을 구하시오
10. 데이터가 할인률이 15%, 온도가 25, 광고비가 300,000원 일때 90%의 신뢰구간과 예측구간을 구하시오.
11. 독립변수가 할인율과 온도를 고정한 상태에서 광고비가 배달 주문량에 영향을 주는지 가설을 검정하시오.
    - 귀무가설 : 광고비의 회귀계수는 0이다.
    - 대립가설 : 광고비의 회귀계수는 0이 아니다.

```python
import pandas as pd

df = pd.DataFrame({
    '할인율':[28,24,13,0,27,30,10,16,6,5,7,11,11,30,25,
           4,7,24,19,21,6,10,26,13,15,6,12,6,20,2],
    '온도':[15,34,15,22,29,30,14,17,28,29,19,19,34,10,
          29,28,12,25,32,28,22,16,30,11,16,18,16,33,12,22],
    '광고비':[342,666,224,764,148,499,711,596,797,484,986,347,146,362,642,
           591,846,260,560,941,469,309,730,305,892,147,887,526,525,884],
    '주문량':[635,958,525,25,607,872,858,732,1082,863,904,686,699,615,893,
           830,856,679,918,951,789,583,988,631,866,549,910,946,647,943]
})
```

```python
# 다중 선형 회귀 모델 적합
from statsmodels.formula.api import ols

model = ols('주문량 ~ 할인율 + 온도 + 광고비',data=df).fit()


# 1. 할인율과 온도의 상관계수를 구하시오
print("1.상관 계수", round(df['할인율'].corr(df['온도']),2))

# 2. 모델의 결정 계수를 구하시오
print("2.결정 계수(R-squre)",round(model.rsquared, 2))

# 3. 각 변수의 회귀 계수를 구하시오
print("3.회귀 계수(기울기)",round(model.params, 4))

# 4. 모델의 절편을 구하시오
print("4.절편",model.params['Intercept'])

# 5. 온도의 회귀계수가 통계적으로 유의한지 검정하시오.
print("5. 회귀 계수 검정",round(model.pvalues['온도']))

# 6. 데이터가 할인율이 10%, 온도가 20, 광고비가 500,000원일때 배달 주문량을 예측하시오.
new_data = pd.DataFrame({"할인율":[10],"온도":[20],"광고비":[500]})
result = model.predict(new_data)
print("6. 신규 데이터 : ", int(result[0]))

# 7. 잔차 제곱합을 구하시오
df['잔차'] = df['주문량'] - model.predict(df)
print("7.잔차 제곱합",round(sum(df['잔차']**2),2))

# 8. MSE를 구하시오.
MSE = (df['잔차']**2).mean()
print('8. MSE : ', round(MSE,4))

# 9. 온도의 회귀 계수에 대한 90% 신뢰 구간을 구하시오
print("9. 신뢰구간:\n", model.conf_int(alpha=0.1))


# 10. 데이터가 할인률이 15%, 온도가 25, 광고비가 300,000원 일때 90%의 신뢰구간과 예측구간을 구하시오.
new_data = pd.DataFrame({'할인율':[15],'온도':[25],"광고비":[300]})
pred = model.get_prediction(new_data)
result = pred.summary_frame(alpha=0.1)

# 11. 독립변수가 할인율과 온도를 고정한 상태에서 광고비가 배달 주문량에 영향을 주는지 가설을 검정하시오.
#     - 귀무가설 : 광고비의 회귀계수는 0이다.
#     - 대립가설 : 광고비의 회귀계수는 0이 아니다.

cond = model.pvalues['광고비']<0.05
if cond:
    result ="기각"
else:
    result ="채택"

print("11. 귀무가설", result)

print(model.summary())
```

## section 09. 로지스틱 회귀


- 주어진 데이터를 사용하여 데이터를 앞 50%의 a와 나머지 50%의 b로 나누기



1. 종속변수 target과 모든 독립변수를 사용하여 로지스틱 회귀 모델을 적합하고, 유의하지 않은 독립변수의 개수를 구하시오.
2. p-value가 0.05보다 작은 유의한 변수만 사용하여 수정된 모델을 만들고 적합하시오. 가장 큰 p-value를 가진 변수의 이름을 구하시오.
3. 수정된 모델에서 독립변수 중 절대값이 가장 큰 회귀 계수를 가진 변수의 이름을 구하시오
4. 수정된 모델에서 로그 우드를 구하시오
5. 수정된 모델에서 잔차이탈도를 구하시오
6. 수정된 모델에서 'booked' 변수가 3증가 할때 오즈비를 계산하시오.
7. 수정된 모델에서 p-value가 0.05보다 작은 회귀 계수의 총합을 구하시오
8. 수정된 모델로 b 데이터를 사용해 예측한 수, b 데이터의 target과 비교해 정확도를 계산하시오.
9. 8에서 계산한 정확도를 바탕으로 오류율을 계산하시오. 

```python
import pandas as pd

df = pd.read_csv('./ch6_data/customer_travel.csv')
df
```

```python
midpoint= len(df)//2
a = df.iloc[:midpoint]
b = df.iloc[midpoint:]

a.shape,b.shape
```

```python
# 1. 유의하지 않은 독립변수의 개수
from statsmodels.formula.api import logit
formula = "target ~ age +service + social + booked"

model = logit(formula,data=a).fit()

print("1.", sum(model.pvalues[1:] >=0.05))

# 2. 수정된 모델에서 가장 큰 p-value를 가진 변수의이름

formula = 'target ~ age + booked'
model = logit(formula, data=a).fit()

print("2. ", model.pvalues[1:].idxmax())

# 3. 수정된 모델에서 독립변수 중 절댓값이 가장 큰 회귀 계수를 가진 변수의 이름
print("3. ", model.params[1:].abs().idxmax())

# 4. 로그 우도
print("4. ", model.llf)

# 5. 잔차 이탈도
print("5." , -2*model.llf)

# 6. 'booked' 변수가 3증가할 때 오즈비
import numpy as np

print("6. ",np.exp(model.params['booked']*3))

# 7. p-value가 0.05보다 작은 회귀계수의 총합
print("7. ",model.params[model.pvalues<0.05].sum())

# 8. 정확도
pred = model.predict(b)
pred = (pred>0.5).astype(int)

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(b['target'],pred)
print("8.", accuracy)

# 9. 오류율

error_rate = 1 - accuracy
print("9. ", error_rate)



```
