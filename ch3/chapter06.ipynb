## Section01 : 단일 표본 검정

### 문제 1.
- 한 커피 제조회사에서는 새로 출시한 커피의 카페인 함량이 평균 95mg 미만으로 주장
- 25개의 커피 샘플을 무작위로 추출하여 커피 제조회사의 주장이 타당한지 검정
- 귀무가설 : u>=95mg
- 대립가설 : u< 95mg


1. 표본 데이터의 평균을 구하시오
2. Shaprio-wilk 검정의 p-value를 구하시오
3. 단일 표본 t-검정의 통계량을 구하시오
4. 단일 표본 t-검정의 p-value를 구하시오
5. 유의수준 0.05이하 귀무가설을 기준으로 검정 결과를 채택/기각 중 선택해 입력


```python
import pandas as pd
df = pd.DataFrame({
    'Caffeine(mg)':[
        94.2,93.7,95.5,93.9,94.0,95.2,94.7,93.5,92.8,94.4,
        93.8,94.6,93.3,95.1,94.3,94.9,93.9,94.8,95.0,94.2,93.7,94.4,95.1,94.0,93.6
    ]
})
```

```python
# 1. 표본 평균
print(df.mean())

# 2. 정규성 검정
from scipy import stats
print(stats.shapiro(df['Caffeine(mg)']))

# 3. 단일 표본 t-검정
print(stats.ttest_1samp(df['Caffeine(mg)'],95,alternative='less'))
```

```python
statistic,pvalue = stats.ttest_1samp(df['Caffeine(mg)'],95,alternative='less')
print("{:.10f}".format(pvalue))
```

```python
# 1. 표본 데이터의 평균을 구하시오
    # -> 94.264
# 2. Shaprio-wilk 검정의 p-value를 구하시오
    # -> 0.9322031137747027
# 3. 단일 표본 t-검정의 통계량을 구하시오
    # ->-5.501737036221897
# 4. 단일 표본 t-검정의 p-value를 구하시오
    # 0.0000058687
# 5. 유의수준 0.05이하 귀무가설을 기준으로 검정 결과를 채택/기각 중 선택해 입력
    # 기각
```

## Section02 독립 표본 검정

### 문제 2. 
- 스마트폰 배터리의 충전 시간을 줄이기 위한 새로운 충전기 개발
- 스마트폰 사용자 집단을 두 그룹으로 나눠 한 그룹에는 새로운 충전기를 제공하고, 다른 그룹에는 기존의 충전기를 제공
- 두 그룹 충전 시간은 정규 분포를 따르며 분산을 같다고 가정
- 새로운 충전기가 더 빠르다고 할 수 있는지를 유의수준 5%이하에서 검정
---
- 귀무가설 : u1>=u2
- 대립가설 : u1<u2

1. 위의 가설을 검정하기 위한 검정 통계량을 구하시오.
2. 위의 통계량에 대한 p-value를 구하시오.
3. 유의수준 0.05하에서 귀무가설을 기준으로 검정의 결과를 채택/기각 중 선택


```python
import pandas as pd

df = pd.DataFrame({
    '충전기':['New']*10 + ['Old']*10,
    '충전시간':[
        1.5,1.6,1.4,1.7,1.5,1.6,1.7,1.4,1.6,1.5,
        1.7,1.8,1.7,1.9,1.8,1.7,1.8,1.9,1.7,1.6
    ]
})
df
```

```python
from scipy import stats

new_cond = df['충전기']=='New'
old_cond = df['충전기']=='Old'

print(stats.ttest_ind(df[new_cond]['충전시간'],df[old_cond]['충전시간'],alternative='less',equal_var=True))
```

```python
# 1. 위의 가설을 검정하기 위한 검정 통계량을 구하시오.
    # -> 4.582575694955849
# 2. 위의 통계량에 대한 p-value를 구하시오.
    # -> 0.00011546547787696304
# 3. 유의수준 0.05하에서 귀무가설을 기준으로 검정의 결과를 채택/기각 중 선택
    # -> 기각

```

## Section04. 일원 분산 분석


## 문제 4.
- 4개의 중학교에서 다른 교육 방법을 사용해 학생들에게 수학을 가르쳤다.
- 각 중학교의 학생들이 받은 교육 후의 수학 성적은 다음과 같아.
- 4개의 교육방법에 따른 성적 차이가 통계적으로 유의미한지 5%의 유의수준에서 검정.
---
- 귀무가설 : 모든 중학교의 학생들의 수학 성정 평균은 동일하다.
- 대립가설 : 적어도 2개 중학교의 학생들의 수학 성적 평균은 다르다.

1. 각 그룹의 수학 성적에 대해 shaprio-wilk 검정을 통한 정규성을 확인, p-value
2. 4개 그룹의 수학 성정이 등분산성을 갖는기 확인하기위해 levence 검정을 실시 p-value
3. 유의수준 0.05하에서 귀무가설을 기준으로 검정의 결과를 채택/기각 중 선택해 입력
4. 그룹 변수의 자유도
5. 잔차의 자유도
6. 성적의 제곱합
7. 성적의 평균 제곱
8. F-통계량
9. 성적에 대한 p-value

```python
import pandas as pd

df = pd.read_csv('./ch6_data/math.csv')
df['groups'].value_counts()
```

```python
from scipy import stats

# 1. shaprio-wilk 검정
condA = df['groups'] == 'group_A'
print("A",stats.shapiro(df[condA]['scores']))

condB = df['groups'] == 'group_B'
print("B",stats.shapiro(df[condB]['scores']))

condC = df['groups'] == 'group_C'
print("C",stats.shapiro(df[condC]['scores']))

condD = df['groups'] == 'group_D'
print("D",stats.shapiro(df[condD]['scores']))
```

```python
# 2. Levence 검정(등분산성)

print(stats.levene(df[condA]['scores'],df[condB]['scores'],df[condC]['scores'],df[condD]['scores']))
```

```python
# 3. 일원분산분석을 위한 모델 학습
from statsmodels.formula.api import ols

model =ols('scores ~ groups',df).fit()
```

```python
# 3~9 ANOVA 테이블
from statsmodels.stats.anova import anova_lm
print(anova_lm(model))
```
