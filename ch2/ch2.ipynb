#### 결측치 채우기(범주형)
- 범주형 데이터에서 결측치를 채울 때는 주로 최빈값으로 대체
- 특정 컬럼에서 mode()[0]을 통해 최빈값을 찾고, fillna()로 값을 채움


```python
m = train['workclass'].mode()[0]
train['workclass'] = train['workclass'].fillna(m)

m = train['native.country'].mode()[0]
train['native.country'] = train['native.country'].fillna(m)

train.isnull().sum()
```

```python
train['occupation']  = train['occupation'].fillna('X')
train.isnull().sum()
```

```python
test['workclass'] = test['workclass'].fillna(test['workclass'].mode()[0])
test['native.country'] = test['native.country'].fillna(test['native.country'].mode()[0])
test['occupation'] = test['occupation'].fillna('X')
```

- 나이가 소수점으로 채워지지 않도록 int() 함수 사용


```python
value = int(train['age'].mean())

print("평균값: ", value)

train['age'] = train['age'].fillna(value)
test['age'] = test['age'].fillna(value)
```

### 검증 데이터 나누기
- 검증 데이터를 나누는 작업도 데이터 전처리 작업중 하나이다.
- 검증 데이터를 나누는 이유는 학습된 모델의 성능을 평가하고 개선하기 위함이다.
- 검증 데이터를 활용해 모델의 성능을 평가하고, 데이터 전처리 단계 또는 하이퍼 파라미터 튜닝을 개선할 수 있다.
- Sklearn에서 제공하는 train_test_split()을 활용해 데이터를 분리한다.
- train과 validation을 나누는 용도로 활용한다.

- train_test_split(train,target,test_size=0.2, random_state=0)
    -  train : X데이터, 데이터 프레임 형태로 입력.
    -  target : label(y) 데이터, 시리즈 형태로 입력한다.
    -  test_size = 0.2 검증용 데이터 비율로 0.2는 전체 데이터 중에서 20%를 의미한다.
    -  random_state = 0 랜덤적인 요소를 고정, 만약 작성하지 않으면 실행 때마다 다른값으로 나뉜다.
    

##### train_test_split()은 반환값이 4개다. X,X,Y,Y로 암기
- X_train, X_val, Y_train, Y_val로 된다.

```python
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val = train_test_split(
    train, y_train,test_size=0.2, random_state=0
)
X_train.shape,X_val.shape,y_train.shape,y_val.shape
```

### 머신러닝 학습 및 평가
- 의사결정 나무 모델은 학습을 통해 다음과 같이 트리구조를 만든다.
- 의사결정 나무는 어떤 과정을 통해 결과가 도출되었는지 설명하기 쉽다.
- 하지만, 과적합 되기 쉬워 성능에 한계가 있다.
- 이를 보완하기 위해 여러 개의 모델을 학습시켜 사용한다.
- -> 앙상블이라고 부르며 , 두 가지 기법이 있다.
- 배깅(bagging), 부스팅(Boosting)
    - 배깅 : 랜덤포레스트(Random Forest)
    - 부스팅 : 라이트지비엠(LightGBM)

#### 랜덤 포레스트
- 여러 개의 의사결정 나무를 기반으로 한 앙상블 학습 알고리즘
- 사이킷런에서 랜덤포레스트 분류 모델을 불러와 3단계로 사용한다.
- 1. rf = RandomForestClassifier() #모델을 선택
- 2. rf.fit(X_train,y_train) #학습을 진행
- 3. pred = rf.predict_proba(X_val) #예측을 한다.

- predict와 predict_proba가 있다.
- predict는 예측된 각 레이블(클래스)을 반환
- predict_proba 는 각 레이블에 속할 확률값을 반환한다.


```python
# 랜덤 포레스트
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)
pred = rf.predict_proba(X_val)

print(rf.classes_)
pred[:10]
```

#### 평가지표
- 머신러닝 모델을 학습하고 나서 제대로 학습이 되었고, 예측을 하고 있는지 검증 데이터로 평가가 필요함.
- 분류 모델의 평가지표에는 여러 가지가 있다.
1. 정확도(Accuracy)
2. 정밀도(Precision)
3. 재현율(Recall)
4. F1 스코어(F1-score)
5. ROC_AUC


##### (1) ROC_AUC
- 평가지표인 ROC_AUC 점수를 평가할 때는 확률값이 필요하다.
- 예측할 때 확률값을 예측할 수 있는 predict_proba()를 사용


```python
from sklearn.metrics import roc_auc_score
roc_auc = roc_auc_score(y_val, pred[:,1])
print('roc_auc',roc_auc)
```

##### (2) Accuracy(정확도)
- 정확도를 평가할 때 predict()를 활용
- predict()를 예측된 각 클래스(레이블)를 반환한다.



```python
pred = rf.predict(X_val)
pred[:10]
```

```python
# Accuracy(정확도)의 결과 점수를 확인해 보자
# 전체 데이터 중 올바르게 예측된 데이터의 비율이다.

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_val, pred)
print('accuracy_score',accuracy)
```

##### (3) F1 Score
- F1 Score는 정밀도와 재현율의 조화 평균으로 계산되는 평가 지표이며 높을 수록 좋다.



```python
from sklearn.metrics import f1_score
f1 = f1_score(y_val, pred,pos_label='>50K')
print('f1_score',f1)
```

### LightGBM
- 그래디언트 부스팅(Gradient Boosting) 기반의 앙상블 학습 알고리즘
- 머신러닝 모델 중 정형 데이터를 다룰 때 매우 인기 있는 모델이다.



#### 파라미터
- random_state : 모델을 학습할 때 랜덤적인 요소가 존재하기 때문에 매번 결과가 달라진다.
- verbose : 로그 메시지를 숨기기위함.


```python
# !pip install lightgbm
```

```python
import lightgbm as lgb
lgbmc = lgb.LGBMClassifier(random_state=0, verbose=1)
lgbmc.fit(X_train,y_train)
pred = lgbmc.predict_proba(X_val)

roc_auc = roc_auc_score(y_val, pred[:,1])
print('roc_auc : ',roc_auc)

pred = lgbmc.predict(X_val)
accuracy = accuracy_score(y_val,pred)
print("accuracy_score : ", accuracy)

f1 = f1_score(y_val,pred,pos_label=">50K")
print("f1_score : ",f1)
```
