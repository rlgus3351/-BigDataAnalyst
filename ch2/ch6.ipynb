## section 1 : 환자의 당뇨병 여부 예측
- 예측할 컬럼 : Outcome(0:정상 , 1: 당뇨병)
- 학습용 데이터를 이용해 환자의 당뇨병을 예측하는 모델을 만든후 이를 평가용 데이터에 적용해 얻은 예측값을 csv 파일로 생성


```python
#1. 문제 정의
# 평가 : roc-auc
# target = Outcome
```

```python
import pandas as pd
train = pd.read_csv('./ch6_data/diabetes_train.csv')
test = pd.read_csv('./ch6_data/diabetes_test.csv')
```

```python
# 탐색적 데이터 분석

print("Train shape", train.shape)
print("Test shape", test.shape)

print("train info", train.info())
print("test info", test.info())
```

```python
print("target 빈도", train['Outcome'].value_counts())
```

```python
# 4. 데이터 전처리
# 타깃 분리
target = train.pop("Outcome")
```

```python
# 5. 검증 데이터 나누기

from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val = train_test_split(
    train,
    target,
    test_size = 0.2,
    random_state = 0
)
X_train.shape,X_val.shape,y_train.shape,y_val.shape 
```

```python
# 6-2. 머신러닝 평가
from sklearn.metrics import roc_auc_score
roc_auc = roc_auc_score(y_val, pred[:,1])
print('roc_auc', roc_auc)
```

```python
# 7. 예측 및 결과 파일 생성
pred = rf.predict_proba(test)
sumbit = pd.DataFrame({'pred':pred[:,1]})
sumbit.to_csv("./ch6_data/section1_result.csv",index=False)
```

## Section 1 : 심화

```python
import pandas as pd

train = pd.read_csv('./ch6_data/diabetes_train.csv')
test = pd.read_csv('./ch6_data/diabetes_test.csv')

target = train.pop('Outcome')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train = scaler.fit_transform(train)
test = scaler.transform(test)
```

```python
from sklearn.model_selection import train_test_split
X_tr,X_val,y_tr,y_val = train_test_split(train,target,test_size=0.2, random_state=0)


# 6. 머신러닝 학습
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=5, n_estimators=500, random_state=0)
rf.fit(X_tr,y_tr)
pred = rf.predict_proba(X_val)

# 6-2. 머신러닝 평가
from sklearn.metrics import roc_auc_score
roc_auc = roc_auc_score(y_val, pred[:,1])
print('roc_auc', roc_auc)

# 7. 예측 및 결과 파일 생성
pred = rf.predict_proba(test)
sumbit = pd.DataFrame({'pred':pred[:,1]})
sumbit.to_csv("./ch6_data/section1_result_2.csv",index=False)
```

### Section 2 : 심화
- 1. 데이터 전처리 
    - 레이블 인코딩 : 베이스라인에 있는 원-핫 인코딩을 레이블 인코딩으로 변경

    - max_depth : 3~7

```python
from sklearn.preprocessing import LabelEncoder

combined = pd.concat([train,test])
cols = train.select_dtypes(include='object').columns
```

```python
for col in cols:
    le = LabelEncoder()
    combined[col] = le.fit_transform(combined[col])
```

## section 3 : 신용카드 신청자의 채무 불이행을 예측하시오.

### 1. 문제정의
- pred : 예측값
- 제출한 모델의 성능은 f1 평가 지표에 따라 채점한다.


```python
import pandas as pd

train = pd.read_csv('./ch6_data/creditcard_train.csv')
test = pd.read_csv('./ch6_data/creditcard_test.csv')
```

```python
print("train : ", train.shape)
print("test : ", test.shape)
```

```python
train.info()
```

```python
test.info()
```

```python
train['STATUS'].value_counts()
```

```python
cols = train.select_dtypes(include='object').columns
for col in cols:
    set_train = set(train[col])
    set_test = set(test[col])
    same =(set_train == set_test)
    if same:
        print(col,"\t카테고리 동일함")
    else:
        print(col,"\t카테고리 동일하지 않음")
```

```python
train.dropna(subset=['OCCUPATION_TYPE'], inplace=True)
```

```python
target = train.pop('STATUS')
```

```python
# 원핫 인코딩
train = pd.get_dummies(train)
test = pd.get_dummies(test)
```

```python
from sklearn.model_selection import train_test_split
X_tr,X_val,y_tr,y_val = train_test_split(
    train,
    target,
    test_size=0.2,
    random_state=0
)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_tr,y_tr)

pred = rf.predict(X_val)
```

```python
from sklearn.metrics import f1_score
score = f1_score(y_val,pred)
print('\n f1:', score)
```

```python
pred = rf.predict(test)
submit = pd.DataFrame({'pred':pred})
submit.to_csv("./ch6_data/section3_result.csv",index=False)
```

### 2. 성능 개선
- 심각한 불균형 데이터일 경우 평가지표가 낮게 나올 수 있다.
- f1_score는 1에 가까울 수록 좋은 성능이다.
- 데이터 전처리
    - 1. 레이블 인코딩 : 베이스라인에 있는 원-핫 인코딩을 레이블 인코딩으로 변환
    - 2. 스케일링 효과가 없음.
    - 3. id 제거로 향상
- 하이퍼 파라미터 튜닝
    - 1. max_depth:3~7
    - 2. n_estimators:200~500
    - 3. class_weight='balanced'

```python
import pandas as pd

train = pd.read_csv('./ch6_data/creditcard_train.csv')
test = pd.read_csv('./ch6_data/creditcard_test.csv')
```

```python
freq = train['OCCUPATION_TYPE'].mode()[0]
```

```python
train['OCCUPATION_TYPE'] = train['OCCUPATION_TYPE'].fillna(freq)
test['OCCUPATION_TYPE'] = test['OCCUPATION_TYPE'].fillna(freq)
target = train.pop("STATUS")
```

```python
train = train.drop('ID',axis=1)
test = test.drop('ID',axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
cols = train.select_dtypes(include='object').columns
for col in cols:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    test[col]= le.transform(test[col])
```

```python
from sklearn.model_selection import train_test_split
X_tr,X_val,y_tr,y_val = train_test_split(
    train,
    target,
    test_size=0.2,
    random_state=0
)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_tr,y_tr)
pred = rf.predict(X_val)
```

```python
from sklearn.metrics import f1_score
score = f1_score(y_val,pred)
print('f1',score)
```

```python
pred = rf.predict(test)
submit = pd.DataFrame({'pred':pred})
submit.to_csv("./ch6_data/section3_result2.csv", index=False)
```
